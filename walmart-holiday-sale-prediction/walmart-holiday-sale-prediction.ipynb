Given *data about weekly Walmart sales*, let's try to predict whether a given record will be from **during a holiday**.

We will use logistic regression, support vector machine, and decision tree classifiers to make our predictions.

Data source: https://www.kaggle.com/datasets/aditya6196/retail-analysis-with-walmart-data

### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('Walmart_Store_sales.csv')
data
```

```python
data.info()
```

### Feature Engineering: Year and Month

```python
data['Year'] = data['Date'].apply(lambda x: x[-4:])
data['Month'] = data['Date'].apply(lambda x: x[3:5])

data = data.drop('Date', axis=1)
```

```python
data
```

#### Visualizing Correlations

```python
corr = data.corr()

plt.figure(figsize=(8,8))
sns.heatmap(corr, annot=True, vmin=-1.0, cmap='mako')
plt.show()
```

#### Encoding Store column

```python
def onehot_encode(df, column, prefix):
    df = df.copy()
    dummies = pd.get_dummies(df[column], prefix=prefix, dtype=int)
    df = pd.concat([df, dummies], axis=1)
    df = df.drop(column, axis=1)
    return df
```

```python
data = onehot_encode(data, 'Store', 'store')
data
```

#### Splitting/Scaling

```python
y = data['Holiday_Flag'].copy()
X = data.drop('Holiday_Flag', axis=1).copy()
```

```python
scaler = StandardScaler()

X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns, index=X.index)
```

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=1)
```

### Modeling/Training

```python
log_model = LogisticRegression()
svm_model = SVC()
dec_model = DecisionTreeClassifier()
```

```python
log_model.fit(X_train, y_train)
svm_model.fit(X_train, y_train)
dec_model.fit(X_train, y_train)

print("Models trained.")
```

### Results

```python
print("Logistic Regression Accuracy:" ,log_model.score(X_test, y_test))
print("      Support Vector Machine:", svm_model.score(X_test, y_test))
print("               Decision Tree:", dec_model.score(X_test, y_test))
```
