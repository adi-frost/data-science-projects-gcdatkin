## Tabular Zoo Animal Classification

Given *data about zoo animals*, we are going to predict the type of animal in the zoo.

We will be using Logistic Regression, Support Vector Machine and Neural Networks to make our predictions.

Data source: https://www.kaggle.com/datasets/uciml/zoo-animal-classification?select=zoo.csv

### Importing Libraries

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
data = pd.read_csv('zoo.csv')
data
```

### Preprocessing

```python
data.drop('animal_name', axis=1, inplace=True)
```

```python
data.info()
```

```python
data.isna().sum().sum()
```

```python
# Splitting and Scaling
y = data['class_type']
X = data.drop('class_type', axis=1)
```

```python
scaler = MinMaxScaler()

X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X
```

```python
X.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=1)
```

### Training

```python
# Logistic Regression
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
```

```python
# Support Vector Machine
svm_model = SVC(C=1.0)
svm_model.fit(X_train, y_train)
```

```python
# Neural Network
nn_model = MLPClassifier(hidden_layer_sizes=(64, 64))
nn_model.fit(X_train, y_train)
```

### Results

```python
log_acc = log_model.score(X_test, y_test)
svm_acc = svm_model.score(X_test, y_test)
nn_acc = nn_model.score(X_test, y_test)

print("Accuracy Results\n" + "*"*20)
print("      Logistic Model: {:.2f}%".format(log_acc*100))
print("           SVM Model: {:.2f}%".format(svm_acc*100))
print("Neural Network Model: {:.2f}%".format(nn_acc*100))
```
