## Candy Bar Prediction

Given *data about attributes of various candies*, we need to predict if a given candy is a bar or not. 

Data source: https://www.kaggle.com/datasets/fivethirtyeight/the-ultimate-halloween-candy-power-ranking

### Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('candy-data.csv')
data
```

### Preprocessing

```python
df = data.copy()
```

```python
df = df.drop('competitorname', axis=1)
```

#### Visualizating

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(), annot=True, vmin=-1, vmax=1)
plt.show()
```

```python
df.isnull().sum()
```

```python
y = df['bar']
X = df.drop('bar', axis=1)
```

```python
scaler = MinMaxScaler()

X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
X
```

```python
# Split X
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=6)
```

### Training

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
print(f"Model Accuracy: {model.score(X_test, y_test)}")
```
