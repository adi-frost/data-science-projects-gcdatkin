## Predicting University Rankings

Data Source: https://www.kaggle.com/datasets/mylesoneill/world-university-rankings

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('cwurData.csv')
data
```

```python
data = data.drop('institution', axis=1)
data
```

```python
data = data.drop('year', axis=1)
data
```

```python
data.isnull().sum()
```

```python
data.drop('broad_impact', axis=1, inplace=True)
```

```python
data
```

```python
encoder = LabelEncoder()

data['country'] = encoder.fit_transform(data['country'])
country_mappings = {index: label for index, label in enumerate(encoder.classes_)}
```

```python
country_mappings
```

```python
data
```

```python
y = data['world_rank']
X = data.drop('world_rank', axis=1)
```

```python
scaler = StandardScaler()

X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, shuffle=True, random_state=64)
X_train.shape, X_test.shape
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print("R Squared: {:.2f}%".format(model.score(X_test, y_test)*100))
```
