Given *EEG data from students durring MOOC lectures*, let's try to predict whether a given student will be **confused** during a lecture.

We will use a Keras/Tensorflow neural network to make our predictions.

Data source: https://www.kaggle.com/datasets/wanghaohan/confused-eeg

### Importing Libraries

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

import tensorflow as tf

from sklearn.metrics import confusion_matrix, classification_report
```

```python
%matplotlib inline
sns.set(style='darkgrid', color_codes=True)
```

```python
eeg_df = pd.read_csv('EEG_data.csv')
info_df = pd.read_csv('demographic_info.csv')
```

```python
eeg_df
```

```python
info_df
```

#### Merging Dataframes

```python
info_df.rename(columns={'subject ID': 'SubjectID'}, inplace=True)
```

```python
data = info_df.merge(eeg_df, on='SubjectID')
data
```

```python
data.info()
```

### Cleaning

```python
df = data.copy()
```

```python
df = df.drop(['SubjectID', 'VideoID', 'predefinedlabel'], axis=1)
```

```python
df.columns
```

```python
df.rename(columns={' age': 'Age', ' ethnicity': 'Ethnicity', ' gender': 'Gender', 'user-definedlabeln': 'Label'}, inplace=True)
```

```python
df['Label'] = df['Label'].astype(int)
df
```

```python
print("Missing values:", df.isna().sum().sum())
```

### Encoding features

```python
df['Gender'].unique()
```

```python
df['Gender'] = df['Gender'].apply(lambda x: 1 if x == 'M' else 0)
```

```python
df['Ethnicity'].unique()
```

```python
ethnicity_dummies = pd.get_dummies(df['Ethnicity'])
df = pd.concat([df, ethnicity_dummies], axis=1)
df = df.drop('Ethnicity', axis=1)
```

```python
df
```

```python
print('Non-numeric columns:', len(df.select_dtypes('object').columns))
```

### EDA

```python
df.dtypes
```

### Splitting/Scaling

```python
y = df['Label'].copy()
X = df.drop('Label', axis=1).copy()
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=123)
```

### Training

```python
inputs = tf.keras.Input(shape=(X_train.shape[1]))
inputs
```

```python
x = tf.keras.layers.Dense(256, activation='relu')(inputs)
x = tf.keras.layers.Dense(256, activation='relu')(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)

model = tf.keras.Model(inputs, outputs)

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics=[
        'accuracy',
        tf.keras.metrics.AUC(name='auc')
    ]
)

batch_size = 32
epochs = 50

history = model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    batch_size=batch_size,
    epochs=epochs,
    callbacks=[
        tf.keras.callbacks.ReduceLROnPlateau()
    ]
)
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(epochs), history.history['loss'], label='Training Loss')
plt.plot(range(epochs), history.history['val_loss'], label='Validation Loss')

plt.xlabel('Epoch')
plt.ylabel("Loss")
plt.title('Loss Over Time')
plt.legend()
plt.show()
```

### Results

```python
model.evaluate(X_test, y_test)
```

```python
y_true = np.array(y_test)

y_pred = np.squeeze(model.predict(X_test))
y_pred = np.array(y_pred >= 0.5, dtype=int)
```

```python
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(4,4))
sns.heatmap(cm, annot=True, fmt='g', vmin=0, cbar=False)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('confusion Matrix')
plt.show()
```

```python
print(classification_report(y_true, y_pred))
```
