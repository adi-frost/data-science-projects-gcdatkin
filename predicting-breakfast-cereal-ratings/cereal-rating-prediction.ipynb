## Cereal Rating Prediction

Given *data about different breakfast cereals*, let's try to predict the **rating** of a given cereal.

We will use a linear regression model to make our predictions.

Data source: https://www.kaggle.com/datasets/crawford/80-cereals

### Importing Libraries

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression, Ridge, Lasso
```

```python
data = pd.read_csv('cereal.csv')
data
```

```python
data.info()
```

### Preprocessing

#### Missing values

```python
(data == -1).sum()
```

```python
data = data.replace(-1, np.NaN)
```

```python
data.isna().sum()
```

```python
for column in ['carbo', 'sugars', 'potass']:
    data[column] = data[column].fillna(data[column].mean())
```

```python
data.isna().sum().sum()
```

```python
data = data.drop('name', axis=1)
```

```python
data
```

#### Encoding

```python
{column: list(data[column].unique()) for column in ['mfr', 'type']}
```

```python
data['type'] = data['type'].apply(lambda x: 1 if x == 'H' else 0)
```

```python
# One-hot encode "mfr" column
dummies = pd.get_dummies(data['mfr'], prefix='mfr', dtype=int)
data = pd.concat([data, dummies], axis=1)
data = data.drop('mfr', axis=1)
```

```python
data
```

#### Splitting and Scaling

```python
y = data.loc[:, 'rating']
X = data.drop('rating', axis=1)
```

```python
X
```

```python
scaler = StandardScaler()

X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
```

### Training

```python
model = LinearRegression()
l2_model = Ridge(alpha=1.0)
l1_model = Lasso(alpha=0.001)
```

```python
model.fit(X_train, y_train)
l1_model.fit(X_train, y_train)
l2_model.fit(X_train, y_train)

print("Models trained.")
```

```python
model_r2 = model.score(X_test, y_test)
l2_model_r2 = l2_model.score(X_test, y_test)
l1_model_r2 = l1_model.score(X_test, y_test)
```

```python
print("R^2 Scores\n" + "*" * 10)
print("Without Regualization: ", model_r2)
print("With L2 (Ridge) Regularization", l2_model_r2)
print("With L1 (Lasso) Regularization", l1_model_r2)
```
